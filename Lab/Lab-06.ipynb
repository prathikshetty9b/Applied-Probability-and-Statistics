```r
#Setup 1
s = c('w1','w2','w3','w4','b1','b2','b3','b4','b5','b6')
p = 1/length(s)*replicate(length(s),1)

nsimulations = 1e6
simulatedData = replicate(nsimulations,sample(s, size = 4, replace = TRUE, prob = p))

checkEvent1 = function(data) {
    if(sum(grepl('w.*',data)) == 2)
    {
        return(1)
    }
    else {
       return(0)
    }
}

mean(apply(simulatedData, 2, checkEvent1))
#j -> Random variable, n-> Sample Size, p-> Success Fraction
dbinom(2,4,4/(4+6))
```

```r
#Setup 2
s = c('w1','w2','w3','w4','b1','b2','b3','b4','b5','b6')
p = 1/length(s)*replicate(length(s),1)

nsimulations = 1e6
simulatedData = replicate(nsimulations,sample(s, size = 4, replace = FALSE, prob = p))

checkEvent2 = function(data) {
    if(sum(grepl('w.*',data)) == 2)
    {
        return(1)
    }
    else {
       return(0)
    }
}

mean(apply(simulatedData, 2, checkEvent2))

dhyper(2,4,6,4)
```

```r
#Setup 3
s = c('w','b')
p = c(0.4,0.6)

nsimulations = 1e6
simulatedData = replicate(nsimulations,sample(s, size = 10, replace = TRUE, prob = p))

checkEvent2 = function(data) {
    if(sum(grepl('w',data)) == 4  & data[10] == 'w')
    {
        return(1)
    }
    else {
       return(0)
    }
}

mean(apply(simulatedData, 2, checkEvent2))
dbinom(,)
```

```r
#Blood Group Problem
#Setup 3
s = c('p','n')
p = c(0.42,0.58)

nsimulations = 1e6
simulatedData = replicate(nsimulations,sample(s, size = 120, replace = TRUE, prob = p))

checkEvent2 = function(data) {
    if(sum(data == 'p') == 30)
    {
        return(1)
    }
    else {
       return(0)
    }
}

mean(apply(simulatedData, 2, checkEvent2))
dbinom(30,120,0.42)
```
