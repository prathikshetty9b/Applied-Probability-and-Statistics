```r
# Number of simulations
nsimulations = 1e4

# Sampling space for host putting the prize
s1 = c('A', 'B', 'C')
p1 = (1/3) * replicate(length(s1), 1)

# Sampling space for participant selecting a door
s2 = c('A', 'B', 'C')
p2 = (1/3) * replicate(length(s2), 1)

# Sampling space for participant switching
s4 = c(0, 1)
p4 = (1/2) * replicate(length(s4), 1)

# Function to simulate one trial of the game
generateOutcome = function(){
  # Array to store the four outcomes: (1) Which door
  # host puts the prize in (2) Which door participant
  # chooses (3) Which door host reveals (4) Whether 
  # participant switches or not
  outcome = replicate(4, 0)
  outcome[1] = sample(s1, size = 1, prob = p1)
  outcome[2] = sample(s2, size = 1, prob = p2)
  # Sanpling space for door opened by host
  if (outcome[1] == outcome[2]){
    s3 = setdiff(c('A', 'B', 'C'), outcome[1])
    p3 = c(1/2, 1/2)
    }
  else {
    s3 = setdiff(c('A', 'B', 'C'), c(outcome[1], outcome[2]))
    p3 = c(1)
   }
  outcome[3] = sample(s3, size = 1, prob = p3) 
  outcome[4] = sample(s4, size = 1, prob = p4)
  return(outcome)
}

simulatedData = replicate(nsimulations, generateOutcome())
#simulatedData

filteredSimulatedData = simulatedData[, simulatedData[4, ] == 1]
#filteredSimulatedData

checkEvent = function(data){
  if (data[1] == data[2]){
    return(0)
  }
  else{
    return(1)
  }
}

mean(apply(filteredSimulatedData, 2, checkEvent))
```

```r
# Load the bats data to a dataframe
dBat = read.csv('bats.csv', header = FALSE, stringsAsFactors = TRUE)
```

```r
# How to access elements of a dataframe?
dBat[1, 2] # access element in the 1st row 2nd column
dBat[2, ] # access all elements in the 2nd row
dBat[, 3] # access all elements in the 3rd column
dBat[1, dBat[1, ] == 'True'] # Gets the genes that are expressed for the 1st bat
```

```r
# Probability that a random bat carries the Ebola virus
mean(dBat[, 6] == 'True')
```

```r
# Probability that gene-1 through gene-5 is expressed in a random bat
for (k in c(1:5)){
  print(mean(dBat[, k] == 'True'))
}  
```

```r
# Is expression of gene-1 through gene-5 dependent or independent
# of the bat carrying the Ebola virus

# Subset the dataframe to those bats that carry the Ebola virus
dBatEbola = dBat[dBat[, 6] == 'True', ]

# Conditional probability that gene-1 through gene-5 is expressed
# given that the bat carries the Ebola virus
for (k in c(1:5)){
  print(mean(dBatEbola[, k] == 'True'))
}  
```
